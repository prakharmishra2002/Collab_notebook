# Exercise 5: Conditional Statements & while loops

1. Separate **positive and negative** cashflows. **Create two new lists!**

```python
#run the cell
cf = [500, 350, 200, 50, -100, -200]
```

```python

cf_pos =[]
cf_neg =[]

for i in cf:
  if i>0:
    cf_pos.append(i)
  else:
    cf_neg.append(i)
print("positive cash flows",cf_pos)
print("negative cash flows",cf_neg)
```

2. Get the **first unprofitable year.** Your Code should return the statement **"Year ... is the first unprofitable Year!"**

```python
for year, cash_flow in enumerate(cf, start=0):
    if cash_flow < 0:
        print(f"Year {year} is the first unprofitable Year!")
        break
```

3. Get the **last profitable year.** Your Code should return the statement **"Year ... is the last profitable Year!"**

```python
last_profitable_year = None

for year in reversed(range(len(cf))):
    if cf[year] > 0:
        last_profitable_year = year
        print(f"Year {year} is the last profitable year!")
        break
```

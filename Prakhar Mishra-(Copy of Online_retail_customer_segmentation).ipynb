## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

````python
'''
Statement 1-  The distribution of 'UnitPrice' in the retail dataset appears to
              be right-skewed, with a majority of products having lower unit
              prices. The KDE curve overlays on the histogram suggests that the
              unit prices tend to concentrate around the lower end, indicating
              that there are more products with relatively lower prices.
              However, there are still a notable number of products with higher
              unit prices, as evidenced by the extended tail on the right side
              of the distribution.

Statement 2-  The pairplot provides insights into the joint distributions and
              correlations between 'Quantity,' 'UnitPrice,' and 'CustomerID'
              variables in the retail dataset. The scatter plots reveal the
              bivariate relationships between each pair of variables, while the
              KDE plots on the diagonal display the univariate distributions for
              each variable.

Statement 3-  The correlation heatmap provides valuable insights into the
              relationships between 'Quantity,' 'UnitPrice,' and 'CustomerID'
              variables in the retail dataset. Each cell in the heatmap
              represents the correlation coefficient, which ranges from -1 to 1,
              indicating the strength and direction of the relationships between
              the variables.'''
````

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.

````python
'''

Null Hypothesis (H0): The distribution of 'UnitPrice' in the retail dataset is
                      not significantly different from a normal distribution,
                      and there is no difference in the concentration of
                      products at different price levels. The KDE curve is
                      merely a result of random variation.

Alternate Hypothesis (H1): The distribution of 'UnitPrice' in the retail dataset
                           is significantly right-skewed, with a higher
                           concentration of products having lower unit prices
                           compared to the normal distribution. The KDE curve
                           represents a genuine pattern in the data, and there
                           is a notable difference in the concentration of
                           products at different price levels.'''
````

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

import scipy.stats as stats

# Extract the 'UnitPrice' data from the retail dataset
unit_price_data = retail['UnitPrice']

# Perform Anderson-Darling test for normality
result = stats.anderson(unit_price_data)

# Extract the p-value from the result
p_value = result[1]

# Print the p-value
print("Anderson-Darling Test p-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Answer Here.

````python
#Anderson-Darling Test
````

##### Why did you choose the specific statistical test?

Answer Here.

````python
'''

The Anderson-Darling test is used to assess whether a sample of data comes from
a specific probability distribution, such as the normal distribution. It is a
variation of the more general Kolmogorov-Smirnov test and is particularly
effective for detecting deviations from normality, especially in the tails of
the distribution.'''
````

### Hypothetical Statement - 2

## ***7. ML Model Implementation***

### ML Model - 1

````python
# ML Model - 1 Implementation

# Fit the Algorithm

# Predict on the model

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
retail = pd.read_csv('/content/drive/MyDrive/Online Retail.csv')

# Assuming 'Quantity', 'UnitPrice', 'CustomerID', and other relevant columns
# Define 'X' (feature matrix)
selected_features = ['Quantity', 'UnitPrice', 'CustomerID']  # Add other relevant columns
X = retail[selected_features]

# Handle missing values in the dataset
X.fillna(0, inplace=True)  # Replace NaN values with 0 or choose an appropriate strategy

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a KMeans model
num_clusters = 5  # Choose the number of clusters based on your problem
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fit the KMeans model
kmeans.fit(X_scaled)

# Assign clusters to each data point
cluster_labels = kmeans.labels_

# Add cluster labels to the original dataset
retail['Cluster'] = cluster_labels

# Visualize the clusters (for 2D visualization)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=cluster_labels, cmap='viridis')
plt.xlabel('Scaled Quantity')
plt.ylabel('Scaled UnitPrice')
plt.title('Customer Segmentation Clusters')
plt.show()

# Explore the characteristics of each cluster
cluster_summary = retail.groupby('Cluster')[selected_features].mean()

# Print the cluster summary
print(cluster_summary)





# New data points to predict

new_data = [[10, 5, 12345],
            [20, 10, 67890]]

# Standardize the new data using the same scaler
new_data_scaled = scaler.transform(new_data)

# Predict the clusters for the new data
predicted_clusters = kmeans.predict(new_data_scaled)

# Print the predicted clusters
for i, cluster in enumerate(predicted_clusters):
    print(f"Data point {i+1} belongs to Cluster {cluster}")
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
# KMeans is an unsupervised learning algorithm used for clustering data points
# into groups, or clusters, based on their similarities. It aims to partition the
# data into K distinct, non-overlapping clusters, where each data point belongs to
# the cluster with the nearest mean (centroid). KMeans is a centroid-based
# algorithm, meaning it calculates the mean of the data points in each cluster and
# uses these means to determine the cluster assignment of new data points.



# Visualizing evaluation Metric Score chart
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# Determine the range of cluster numbers to consider
max_clusters = 10
inertia_values = []

for k in range(1, max_clusters+1):
    kmeans = KMeans(n_clusters=k,n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Plot the Elbow Method chart
plt.plot(range(1, max_clusters+1), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Cluster Number')
plt.xticks(range(1, max_clusters+1))
plt.show()




````

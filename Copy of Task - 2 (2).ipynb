```python
# run the cell
cf = np.array([-500, 50, 80, 120, 150, 200])
cf
```

```python
r=0.075
discount_factors=1/(1+r)**np.arange(len(cf))
discounted_cashflows=cf*discount_factors
npv = np.sum(discounted_cashflows)
if npv > 0:
    decision = "ABC should pursue the project."
elif npv < 0:
    decision = "ABC should not pursue the project."
print("NPV:", npv)
print("Decision:", decision)
```

## Linear Regression with scikit-learn - a simple Introduction

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# lets take an example we have numpy array budget
budget = np.array([5, 10, 17, 27, 35, 40, 42, 49, 54, 60])
```

```python
#numpy array revenue
revenue = np.array([2.6, 19. , 23.8, 26.9, 41.1, 58.3, 40.3, 58.7, 73.1, 69.7])
```

```python
#Create a dataframe using budget and revenue column
df = pd.DataFrame({'budget':budget, 'revenue':revenue})
df
```

```python
plt.scatter(df['budget'], df['revenue'])

plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Linear Relationship between Budget and Revenue')

plt.show()
```

image.png

```python
#import linear regression from scikit learn
from sklearn.linear_model import LinearRegression
```

```python
#Create linear regression object and Fit intercept
X = budget.reshape(-1, 1)
y = revenue
model = LinearRegression(fit_intercept=True)

model.fit(X, y)

print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_)
```

```python
#Fit the model(Regression line)
y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label='Data')


plt.plot(X, y_pred, color='red', label='Regression Line')

plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Linear Regression: Budget vs Revenue')


plt.legend()


plt.show()
```

```python
#find slope of regression line
slope = model.coef_[0]
slope
```

```python
#find intercept of regression line
intercept = model.intercept_
intercept
```

```python
#predict the revenue on budget
new_budget = np.array([1800000, 2600000, 3200000]).reshape(-1, 1)
predicted_revenue = model.predict(new_budget)


for i in range(len(new_budget)):
    print("For budget $", new_budget[i][0], ", the predicted revenue is $", predicted_revenue[i])

```

```python
#Lets create two data points
x_lin = np.array([0, 100])
y_lin = intercept + slope * x_lin
```

```python
#Plot data points in scatter plot and also add a straight line with two data points
plt.scatter(budget, revenue, color='blue', label='Data')

plt.plot(x_lin, y_lin, color='red', label='Straight Line')

plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Scatter Plot with Straight Line')


plt.legend()


plt.show()

```

# Do Multiple Regression Model to predict Financial Returns

**Adding more Features - a multiple Regression Model**

```python
#import five_minute.csv file
#new = pd.read_csv("/content/drive/MyDrive/five_minute.csv")
#new
```

```python
data = pd.read_csv("/content/drive/MyDrive/five_minute.csv", parse_dates = ["time"], index_col = "time")
data
```

```python
data.dropna(inplace = True)
```

```python
#run the cell
data['returns'] = np.log(data.div(data.shift(1)))
data
```

```python
lags = 5
```

```python
#run the cell
cols = []
for lag in range(1, lags + 1):
    col = "lag{}".format(lag)
    data[col] = data.returns.shift(lag)
    cols.append(col)
data.dropna(inplace = True)
```

```python
lm = LinearRegression(fit_intercept = True)
lm
```

```python
lm.fit(data[cols], data.returns)
```

```python
data["pred"] = lm.predict(data[cols].values)
```

```python
data.pred = np.sign(data.pred)
```

```python
data
```

# Perform  In-Sample Backtesting and the Look-ahead-bias and write analysis

```python
#run the cell
data
```

```python
#write strategy in dataframe


# Create a new column to store strategy signals/results
data['strategy'] = None


#If the current day's return is positive, set the 'strategy' column to 'Buy'.
data.loc[data['returns'] > 0, 'strategy'] = 'Buy'

# If the current day's return is negative, set the 'strategy' column to 'Sell'.
data.loc[data['returns'] < 0, 'strategy'] = 'Sell'

# Print the updated DataFrame
print(data)
```

```python
#write cumulative return and cumulative strategy

# Calculate the cumulative return
data['cumulative_return'] = (1 + data['returns']).cumprod()

# Calculate the cumulative strategy performance
data['cumulative_strategy'] = (1 + data['returns'].where(data['strategy'].notnull(), 0)).cumprod()


print(data)
```

```python
#plot the graph using datapoints

plt.plot(data['price'], data['returns'], marker='o', linestyle='-', color='blue')

# Set plot labels and title
plt.xlabel('price')
plt.ylabel('return')
plt.title('Graph of price and return')

# Display the plot
plt.show()
```

```python
#run the cell
data["trades"] = data['pred'].diff().fillna(0).abs()
```

```python
#run the cell
data.trades.value_counts()
```

```python
#What is the highest revenue achieved during the given period? Identify the month/year with the lowest revenue

highest_revenue = df['revenue'].max()
lowest_revenue = df['revenue'].min()
lowest_month_year = df.loc[df['revenue'].idxmin()]
print("Highest Revenue:", highest_revenue)
print("Lowest Revenue:", lowest_revenue)
print("Month/Year with Lowest Revenue:", lowest_month_year)
```

```python
# Calculate the average percentage achieved (revenue as a percentage of the budget) for the dataset.


# Calculate the percentage achieved for each row
df['Percentage Achieved'] = (df['revenue'] / df['budget']) * 100

# Calculate the average percentage achieved
average_percentage_achieved = df['Percentage Achieved'].mean()

# Print the average percentage achieved
print("Average Percentage Achieved:", average_percentage_achieved)
```

```python
#Based on the linear regression model, if the budget is $50, what is the predicted revenue?

budget = 50

# Predict the revenue using the linear regression equation: revenue = intercept + slope * budget
predicted_revenue = intercept + slope * budget

# Print the predicted revenue
print("Predicted Revenue:", predicted_revenue)
```

```python
#Is there a positive or negative relationship between the budget and revenue? Explain.


'''If the slope coefficient is positive, it indicates a positive relationship
 between the budget and revenue. This means that as the budget increases,
 the predicted revenue also increases. On the other hand, if the slope
 coefficient is negative, it suggests a negative relationship, implying that as
 the budget increases, the predicted revenue decreases.'''


```

```python
#Calculate the correlation coefficient between the budget and revenue. Interpret the value.

# Calculate the correlation coefficient between 'Budget' and 'Revenue'
correlation_coefficient = df['budget'].corr(df['revenue'])

# Print the correlation coefficient
print("Correlation Coefficient:", correlation_coefficient)


'''If the correlation coefficient is close to -1, it suggests a strong negative
   linear relationship between the budget and revenue. As the budget increases,
   the revenue tends to decrease.

   If the correlation coefficient is close to 1, it indicates a strong positive
   linear relationship between the budget and revenue. As the budget increases,
   the revenue tends to increase.

   If the correlation coefficient is close to 0, it implies a weak or no linear
   relationship between the budget and revenue. There may be little to no
   association between the two variables.'''
```

```python
#Identify any outliers in the dataset and explain their impact on the linear regression model.



from scipy import stats
# Calculate the Z-score for 'Budget' and 'Revenue'
df['Budget_zscore'] = np.abs(stats.zscore(df['budget']))
df['Revenue_zscore'] = np.abs(stats.zscore(df['revenue']))

# Set a threshold for outlier detection
threshold = 3

# Identify the outliers based on the Z-score
outliers = df[(df['Budget_zscore'] > threshold) | (df['Revenue_zscore'] > threshold)]

# Print the identified outliers
print("Outliers:")
print(outliers)
```

```python
#Suppose the budget for the next period is $70. Estimate the revenue using the linear regression model.


budget_next_period = 70

# Estimate the revenue using the linear regression equation: revenue = intercept + slope * budget_next_period
estimated_revenue = intercept + slope * budget_next_period

# Print the estimated revenue
print("Estimated Revenue:", estimated_revenue)
```

```python
#Compare the actual revenue with the predicted revenue for each entry. Calculate the residuals.


X = df['budget'].values.reshape(-1, 1)
y = df['revenue'].values

# Create a linear regression model
model = LinearRegression()

# Fit the linear regression model to the data
model.fit(X, y)

# Get the intercept and slope coefficients
intercept = model.intercept_
slope = model.coef_[0]

# Calculate the predicted value for each revenue entry
df['Predicted Revenue'] = intercept + slope * 70

# Calculate the residuals (difference between actual and predicted revenue)
df['Residuals'] = df['revenue'] - df['Predicted Revenue']

# Print the DataFrame with actual revenue, predicted revenue, and residuals
print(df[['revenue', 'Predicted Revenue', 'Residuals']])
```
